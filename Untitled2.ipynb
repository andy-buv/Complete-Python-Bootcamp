```python
class Triangle(object):
    """Create a triangle object."""
       
    def __init__(self, t1, t2, t3):
        self.coor1 = t1
        self.coor2 = t2
        self.coor3 = t3
        """
        Heron's Formula for area of a Triangle
        
        area = | Ax(By-Cy)+Bx(Cy-Ay)+Cx(Ay-By) |
               |-------------------------------|
               |               2               | 
        """
        self.area = (((self.coor1[0]*(self.coor2[1]-self.coor3[1])
                      +self.coor2[0]*(self.coor3[1]-self.coor1[1])
                      +self.coor3[0]*(self.coor1[1]-self.coor2[1])
                      )/2.0)**2)**0.5
        
        self.perimeter = 0
        
    def update(self, t1, t2, t3):
        self = Triangle(t1, t2, t3)
    
    def __str__(self):
        return 'Triangle {}: \n A({},{})\n B({},{})\n C({},{})'.format(self, self.coor1[0],self.coor1[1],
                                                                     self.coor2[0],self.coor2[1],
                                                                     self.coor3[0],self.coor3[1]) 
```

```python
a = Triangle((1,1),(1,2),(0,1))
```
